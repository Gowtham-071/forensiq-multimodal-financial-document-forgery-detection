```python
import cv2
import os

input_dirs = {
    "genuine": r"G:\fraud_document_ai\data\processed\grayscale\genuine",
    "fraud": r"G:\fraud_document_ai\data\processed\grayscale\fraud"
}

output_base = r"G:\fraud_document_ai\data\processed\denoised"

for label, in_dir in input_dirs.items():
    out_dir = os.path.join(output_base, label)
    os.makedirs(out_dir, exist_ok=True)

    for file in os.listdir(in_dir):
        img = cv2.imread(os.path.join(in_dir, file), cv2.IMREAD_GRAYSCALE)
        denoised = cv2.GaussianBlur(img, (5, 5), 0)
        cv2.imwrite(os.path.join(out_dir, file), denoised)

print("Denoising complete.")
```
