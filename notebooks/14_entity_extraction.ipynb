```python
import cv2

img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
print("Image loaded:", img is not None)
```

```python
from PIL import Image
import pytesseract

img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

text = pytesseract.image_to_string(Image.fromarray(img_rgb))
print(text)
```

```python
def normalize_numbers(text):
    text = text.replace(" ", "")
    text = text.replace(",", ".")
    return text

norm_text = normalize_numbers(text)
print(norm_text)
```

```python
import re

# VAT percentage
vat_percent = None
vat_match = re.search(r'VAT\[%\](\d+\.?\d*)', norm_text)
if vat_match:
    vat_percent = float(vat_match.group(1))

# Net subtotal
net_subtotal = None
net_match = re.search(r'Networth(\d+\.?\d*)', norm_text)
if net_match:
    net_subtotal = float(net_match.group(1))

# VAT amount
vat_amount = None
vat_amt_match = re.search(r'VAT(\d+\.\d+)', norm_text)
if vat_amt_match:
    vat_amount = float(vat_amt_match.group(1))

# Gross / total amount
gross_total = None
gross_match = re.search(r'Total\$?(\d+\.\d+)', norm_text)
if gross_match:
    gross_total = float(gross_match.group(1))

vat_percent, net_subtotal, vat_amount, gross_total
```

```python
summary_start = norm_text.find("SUMMARY")
summary_text = norm_text[summary_start:]
print(summary_text)
```

```python
vat_percent = None
vat_match = re.search(r'(\d+)%', summary_text)
if vat_match:
    vat_percent = float(vat_match.group(1))

vat_percent
```

```python
numbers = re.findall(r'\d+\.\d+', summary_text)
numbers
```

```python
net_subtotal = float(numbers[0])
vat_amount = float(numbers[1])
gross_total = float(numbers[2])

net_subtotal, vat_amount, gross_total
```

```python
summary_lines = summary_text.split("\n")
for line in summary_lines:
    print(line)
```

```python
total_line = [l for l in summary_lines if "Total" in l][0]
print(total_line)
```

```python
total_numbers = re.findall(r'\d+\.\d+', total_line)
total_numbers
```

```python
net_subtotal = float(total_numbers[0])
vat_amount = float(total_numbers[1])
gross_total = float(total_numbers[2])

net_subtotal, vat_amount, gross_total
```

```python
computed_vat = round(net_subtotal * vat_percent / 100, 2)
computed_vat
```

```python
computed_total = round(net_subtotal + computed_vat, 2)
computed_total
```

```python
vat_diff = abs(vat_amount - computed_vat)
total_diff = abs(gross_total - computed_total)

vat_diff, total_diff
```

```python
if vat_diff > 1 or total_diff > 1:
    decision = "FRAUD"
else:
    decision = "GENUINE"

decision
```

```python
def numeric_fraud_check(net_subtotal, vat_percent, vat_amount, gross_total, tol=1.0):
    computed_vat = round(net_subtotal * vat_percent / 100, 2)
    computed_total = round(net_subtotal + computed_vat, 2)

    vat_diff = abs(vat_amount - computed_vat)
    total_diff = abs(gross_total - computed_total)

    reasons = []

    if vat_diff > tol:
        reasons.append(f"VAT mismatch (expected {computed_vat}, found {vat_amount})")

    if total_diff > tol:
        reasons.append(f"Total mismatch (expected {computed_total}, found {gross_total})")

    if reasons:
        return "FRAUD", reasons
    else:
        return "GENUINE", ["All financial values are consistent"]
```

```python
decision, reasons = numeric_fraud_check(
    net_subtotal,
    vat_percent,
    vat_amount,
    gross_total
)

decision, reasons
```

```python
def interpret_cnn_score(score):
    if score >= 0.65:
        return "High visual tampering risk"
    elif score >= 0.40:
        return "Moderate visual anomalies detected"
    else:
        return "Low visual tampering risk"
```

```python
cnn_score = 0.52  # example value from your model
cnn_reason = interpret_cnn_score(cnn_score)

cnn_score, cnn_reason
```

```python
final_output = {
    "Final Decision": decision,
    "Financial Reasoning": reasons,
    "CNN Risk Score": round(cnn_score, 2),
    "Visual Analysis": cnn_reason
}

final_output
```
