```python
import cv2
import os

input_dirs = {
    "genuine": r"G:\fraud_document_ai\data\raw\genuine",
    "fraud": r"G:\fraud_document_ai\data\raw\fraud"
}

output_base = r"G:\fraud_document_ai\data\processed\grayscale"

for label, in_dir in input_dirs.items():
    out_dir = os.path.join(output_base, label)
    os.makedirs(out_dir, exist_ok=True)

    for file in os.listdir(in_dir):
        img = cv2.imread(os.path.join(in_dir, file))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        cv2.imwrite(os.path.join(out_dir, file), gray)

print("Grayscale conversion complete.")
```
