```python
import os
import cv2

base_input = r"G:\fraud_document_ai\data\processed\thresholded"
base_output = r"G:\fraud_document_ai\data\processed\roi"

os.makedirs(base_output, exist_ok=True)

for label in ["genuine", "fraud"]:
    in_dir = os.path.join(base_input, label)
    out_dir = os.path.join(base_output, label)
    os.makedirs(out_dir, exist_ok=True)

    for file in os.listdir(in_dir):
        img = cv2.imread(os.path.join(in_dir, file), cv2.IMREAD_GRAYSCALE)
        h, w = img.shape

        roi = img[int(h*0.7):h, 0:w]

        cv2.imwrite(os.path.join(out_dir, file), roi)

print("ROI extraction complete.")
```
