```python
import cv2
import pytesseract
import os

input_dirs = {
    "genuine": r"G:\fraud_document_ai\data\processed\thresholded\genuine",
    "fraud": r"G:\fraud_document_ai\data\processed\thresholded\fraud"
}

output_base = r"G:\fraud_document_ai\reports\ocr_text"
os.makedirs(output_base, exist_ok=True)

for label, in_dir in input_dirs.items():
    out_file = os.path.join(output_base, f"{label}_ocr.txt")

    with open(out_file, "w", encoding="utf-8") as f:
        for file in os.listdir(in_dir):
            img = cv2.imread(os.path.join(in_dir, file), cv2.IMREAD_GRAYSCALE)
            text = pytesseract.image_to_string(img)
            f.write(f"\n--- {file} ---\n")
            f.write(text)

print("OCR extraction complete.")
```
