```python
import os
import re

ocr_dir = r"G:\fraud_document_ai\reports\ocr_text"

def has_valid_total(text):
    if "Total" not in text:
        return False
    amounts = re.findall(r"\d+[.,]\d+", text)
    return len(amounts) >= 1

for label in ["genuine", "fraud"]:
    file_path = os.path.join(ocr_dir, f"{label}_ocr.txt")
    flagged = 0
    total = 0

    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()
        docs = content.split("---")

        for doc in docs:
            if doc.strip() == "":
                continue
            total += 1
            if not has_valid_total(doc):
                flagged += 1

    print(f"{label.upper()} → Flagged {flagged} out of {total}")
```

**The system performs OCR-based validation by checking the presence and integrity of critical financial fields such as total amount, flagging documents with missing or malformed values.**
