```python
import cv2
import os

input_dirs = {
    "genuine": r"G:\fraud_document_ai\data\processed\denoised\genuine",
    "fraud": r"G:\fraud_document_ai\data\processed\denoised\fraud"
}

output_base = r"G:\fraud_document_ai\data\processed\thresholded"

for label, in_dir in input_dirs.items():
    out_dir = os.path.join(output_base, label)
    os.makedirs(out_dir, exist_ok=True)

    for file in os.listdir(in_dir):
        img = cv2.imread(os.path.join(in_dir, file), cv2.IMREAD_GRAYSCALE)
        thresh = cv2.adaptiveThreshold(
            img,
            255,
            cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
            cv2.THRESH_BINARY,
            11,
            2
        )
        cv2.imwrite(os.path.join(out_dir, file), thresh)

print("Thresholding complete.")
```
